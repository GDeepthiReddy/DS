### importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

### reading the data

```python
univ = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/Universities.csv")
univ.head()
```

### scalling

```python
# scale down the data
normalized_data = scale(univ.iloc[:,1:])
normalized_data
```

```python
# Dendrogram
from scipy.cluster.hierarchy import linkage
a = linkage(normalized_data,method='complete',metric = 'euclidean')
```

### Dendrogram based on complete linkage

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(25,10))
sch.dendrogram(a)
plt.show()
```

### Applying clustering

```python
# 5 clusters
from sklearn.cluster import AgglomerativeClustering
clust = AgglomerativeClustering(n_clusters = 5,linkage='complete',affinity='euclidean').fit(normalized_data)
```

```python
# cluster labels
clust.labels_
```

```python
# Dataframe for cluster labels
cluster = pd.DataFrame(clust.labels_)
cluster
```

#### creating a new column cluster in dataset

```python
# create a new column 'cluster label' in Univ Dataset
univ['cluster'] = cluster
univ.head()
```

#### Rearranging the columns in a specified order

```python
univ1 = univ.iloc[:,[7,0,1,2,3,4,5,6]]
univ1
```

#### Aggregating the cluster mean

```python
univ1.iloc[:,2:].groupby(univ1.cluster).mean()
```

#### Elbow Method:

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,6):
    Kmeans = KMeans(n_clusters=i, init= 'k-means++', random_state=0)
    Kmeans.fit(normalized_data)
    wcss.append(Kmeans.inertia_)
    
wcss
```

#### optimum number of clusters

```python
plt.plot(range(1,6),wcss)
plt.title('ELbow Method')
plt.xlabel('number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
